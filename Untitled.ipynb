```python
df['total_premiums_paid'] = (df['policy_annual_premium']/12) * df['months_as_customer']
```

```python
df.head()
```

```python
df['net_value_of_customer'] = df['total_premiums_paid'] - df['total_claim_amount']
```

```python
df.head()
```

```python
df.to_csv('Advanced Features Claims Data.csv')
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 25, 35, 50, float('inf')], labels=['Young Adult', 'Adult', 'Middle-Aged', 'Senior'])
```

```python
# Policy State Counts
policy_state_counts = df['policy_state'].value_counts().reset_index()
policy_state_counts.columns = ['policy_state', 'policy_state_count']
```

```python
# Policy CSL Split
df['policy_csl_bodily_injury'], df['policy_csl_property_damage'] = zip(*df['policy_csl'].str.split('/'))
```

```python
# Policy Deductible Categories
df['policy_deductible_category'] = pd.cut(df['policy_deductable'], bins=[0, 1000, 1500, 2000, float('inf')], labels=['<1000', '1000-1500', '1500-2000', '>2000'])
```

```python
# Policy Year
df['policy_year'] = pd.to_datetime(df['policy_bind_date']).dt.year
```

```python
# Auto Year Groups
df['auto_year_group'] = pd.cut(df['auto_year'], bins=[0, 2000, 2010, 2020, float('inf')], labels=['<2000', '2000-2010', '2010-2020', '>2020'])
```

```python
# Fraud Reported Count
fraud_reported_counts = df['fraud_reported'].value_counts().reset_index()
fraud_reported_counts.columns = ['fraud_reported', 'fraud_reported_count']
```

```python
# Save the modified DataFrame
df.to_csv("processed_dataset.csv", index=False)
```
